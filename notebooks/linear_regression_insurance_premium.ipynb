# Introduction


---


This analysis performs a simple linear regression to model the relationship between age and insurance premium. The goal is to predict the insurance premium based on a person's age. We will train a linear regression model on the provided dataset, evaluate its performance, and visualize the results.


---



## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Import Dataset

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/insurance_data.csv')
df
```

## Features and Target

```python
X = df[['Age']]
y = df['Premium']
```

## Train the Model

```python
model = LinearRegression()
model.fit(X, y)
```

## Make Predictions

```python
y_pred = model.predict(X)
```

## Model Parameters

```python
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```

## Model Evaluation

```python
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

## Test the Model

```python
age_to_predict = 30
predicted_premium = model.predict([[age_to_predict]])[0]

print(f"Predicted premium for age {age_to_predict}: {predicted_premium:.2f}")
```

## Visualisation

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.title('Simple Linear Regression: Age vs Premium')
plt.legend()
plt.grid(True)
plt.show()
```

# Conclusion

<p align="justify">The simple linear regression model developed in this analysis provides a clear understanding of the relationship between age and insurance premium based on the provided dataset. The model effectively predicts the insurance premium based on a person's age, as demonstrated by the evaluation metrics and the visualization. The intercept and coefficient values highlight the baseline premium and the linear increase in premium with age. This analysis confirms that age is a significant factor in determining insurance premiums in this dataset and the linear regression model provides a useful tool for prediction.
